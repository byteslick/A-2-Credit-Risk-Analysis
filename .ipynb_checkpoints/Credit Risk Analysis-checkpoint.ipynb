#### loan terms

```python
df.term.value_counts()
```

```python
def change_loan_term(months):
    values = months.split(' ')
    # once we split on blank space and print the result we find that the months have blank space
    # and the result is like: ['', '36', 'months']
    # hence we return second object in the array, that we get after splitting.
    return values[1]
df.term = df.term.apply(lambda x: change_loan_term(x)).astype(int)
df.term.describe()
```
